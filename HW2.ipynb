```python
"""Parse html and return html tables of wikipedia coronavirus_pandemic data."""

#step1: look at html source!"
#step2: get all tables"
#step3: get table we care about"
    
###YOUR CODE HERE###
#step1 : look at html source
url = 'https://en.wikipedia.org/wiki/2019%E2%80%9320_coronavirus_pandemic_by_country_and_territory'
res = requests.get(url)
bs = BeautifulSoup(res.text, 'html.parser')
```

```python
print("step1 : look at html source")
print(bs.prettify())
```

```python
#step2 : get all tables
tables = bs.find_all('table',{'class':'wikitable'})
print("step2 : get all tables")
for table in tables :
    print(table.prettify())
    print("--------------------------")
```

```python
#step3 : get table we care about
covid_table = tables[0]
print("step3 : get table we care about")
print(covid_table.prettify())
```

## 2. Extracting data and filling it into a dictionary(Python data stucture)

#### Extract COVID-19 data for countries from table and store it in dictionary.
*Please print parsed dictionary at the end of code*

##### Shape of Dictionary
- { 'country_name': { 'cases' : 'num', 'deaths' : 'num', 'recovers' : 'num' }}
- *Dictionary data structure does not have order between Keys, so it may show up in different order(index).  You don't need to be worried of it*
- e.g. 'Uzbekistan': {'cases': '2,017', 'deaths': '9', 'recovers': '1,096'}, 'Puerto Rico': {'cases': '1,539', 'deaths': '92', 'recovers': '—'} ...

```python
import re
```

```python
# extract headings and actual data and combine data into single dict
###YOUR CODE HERE###

def find_nation(href):
    return href and re.compile("/wiki/COVID-19_pandemic_in_").search(href)

ctr_list = [c_t.text for c_t in covid_table.find_all(href = find_nation)]
num_list = []
data_list = []

for n in covid_table.find_all('td'):
    if not n.sup:
        num_list.append(n.text.replace('\n',''))

for i in range(0, len(num_list)-3, 3):
    data_list.append({'cases':num_list[i], 'deaths':num_list[i+1], 'recovers':num_list[i+2]})
    
table_dict = {i:j for i,j in zip(ctr_list, data_list)}
print(table_dict)
```

```python
#step1: calculate death rates(to calculate death rates, strings in dataframe should be converted to numeric datas)
#step2: add "death_rates" column to dataframe 
#step3: sort based on death rates 

###YOUR CODE HERE###
for col in df.columns :
    df[col] = df[col].str.replace(',','')
    df[col] = df[col].astype(int)
df['death_rates'] = round(df['deaths'] / df['cases'], 2)
df.sort_values('death_rates', ascending = False)
```

```python
###YOUR CODE HERE###
# highest 5
print("Top 5 countries with the highest death rates by COVID-19")
print(df.sort_values('death_rates', ascending = False).head(5))
for c in df.sort_values('death_rates', ascending = False).head(5).index:
    print(c, end = ' | ')
print("")
print("--------------------------------------------------------")
# lowest 5
print("Top 5 countries with the lowest death rates by COVID-19")
print(df.sort_values('death_rates').head(5))
for c in df.sort_values('death_rates').head(5).index:
    print(c, end = ' | ')
```
